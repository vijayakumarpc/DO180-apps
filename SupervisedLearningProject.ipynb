##### **Part 1.5 a - Experiment with various parameters to improve performance of the base model**

###### **Method-1**

````python
# creating odd list of K for KNN
myList = list(range(3,15))

# subsetting just the odd ones
neighbors = list(filter(lambda x: x % 2 != 0, myList))

#print(myList)
print("KNN neighbors -->",neighbors)
````

````python
# empty list that will hold accuracy scores
ac_scores = []

# perform accuracy metrics for values from 3,5....13
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    # predict the response
    y_pred = knn.predict(X_test)
    # evaluate accuracy
    scores = accuracy_score(y_test, y_pred)
    print("K=", k)
    print(classification_report(y_test,y_pred))

    ac_scores.append(scores)

# changing to misclassification error
MSE = [1 - x for x in ac_scores]

# determining best k
optimal_k = neighbors[MSE.index(min(MSE))]
print("The optimal number of neighbors is %d" % optimal_k)
````

From the classification report, it can be seen that the model has an average performance of around 87% ranging from precision, recall, f1-score, and support. **Accuracy also shows in value of 87% with optimal no.of neighbors set to 9 using the default measurement method 'Euclidean distance'**

###### **Method-2**
**Let's try to use Hyperparameter Tuning to Improve Model Performance.**

````python
#List Hyperparameters that we want to tune.
leaf_size = list(range(1,50))
n_neighbors = list(range(1,30))
p=[1,2]

#Convert to dictionary
hyperparameters = dict(leaf_size=leaf_size, n_neighbors=n_neighbors, p=p)

#Create new KNN object
knn_2 = KNeighborsClassifier()

#Use GridSearch
clf = GridSearchCV(knn_2, hyperparameters, cv=10)

#Fit the model
best_model = clf.fit(X,y)

#Print The value of best Hyperparameters
print('Best leaf_size:', best_model.best_estimator_.get_params()['leaf_size'])
print('Best p:', best_model.best_estimator_.get_params()['p'])
print('Best n_neighbors:', best_model.best_estimator_.get_params()['n_neighbors'])
````

**From GridSearch, it can be seen that the best number of leaf_size is 1 while the optimal distance method is Euclidean or p = 2.**

**Then the most optimal number of K is 25.**

````python
# instantiate learning model (k = 3)
knn = KNeighborsClassifier(n_neighbors = 25,leaf_size=1, p=2)

# fitting the model
knn.fit(X_train, y_train)

# predict the response
y_pred = knn.predict(X_test)

print(classification_report(y_test,y_pred))
````

###### **Method-3**

Let's try another method by specifying the metrics, weights

````python
#Parameter tuning with GridSearchCV

#######################
### K-Nearest Neighbors
#######################
estimator_KNN = KNeighborsClassifier(algorithm='auto')
parameters_KNN = {
    'n_neighbors': (1,23, 1),
    'leaf_size': (20,40,1),
    'p': (1,2),
    'weights': ('uniform', 'distance'),
    'metric': ('minkowski', 'chebyshev')}

# with GridSearch
grid_search_KNN = GridSearchCV(
    estimator=estimator_KNN,
    param_grid=parameters_KNN,
    scoring = 'accuracy',
    n_jobs = -1,
    cv = 5
)
````

````python
from sklearn import metrics
KNN_1=grid_search_KNN.fit(X_train, y_train)
y_pred_KNN1 =KNN_1.predict(X_test)

#Parameter setting that gave the best results on the hold out data.
print(grid_search_KNN.best_params_ )
#Mean cross-validated score of the best_estimator
print('Best Score - KNN:', grid_search_KNN.best_score_ )

print(classification_report(y_test,y_pred_KNN1))
````

````python
# creating odd list of K for KNN
myList = list(range(3,17))

# subsetting just the odd ones
neighbors = list(filter(lambda x: x % 2 != 0, myList))

#print(myList)
print("KNN neighbors -->",neighbors)
````

````python
# empty list that will hold accuracy scores
ac_scores = []

# perform accuracy metrics for values from 3,5....15
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    # predict the response
    y_pred = knn.predict(X_test)
    # evaluate accuracy
    scores = accuracy_score(y_test, y_pred)
    print("K=", k)
    print(classification_report(y_test,y_pred))

    ac_scores.append(scores)

# changing to misclassification error
MSE = [1 - x for x in ac_scores]

# determining best k
optimal_k = neighbors[MSE.index(min(MSE))]
print("The optimal number of neighbors is %d" % optimal_k)
````

**GridSearchCV is used**

````python
from sklearn import svm
#Grid Search CV for SVM
def svc_param_selection(X, y, nfolds):
    Cs = [0.001, 0.01, 0.1, 1, 10]
    gammas = [0.001, 0.01, 0.1, 1]
    my_kernal =['linear','sigmoid','rbf']
    param_grid = {'C': Cs, 'gamma' : gammas, 'kernel': my_kernal}
    grid_search = GridSearchCV(svm.SVC(), param_grid, cv=nfolds)
    grid_search.fit(X, y)
    return grid_search.best_params_


````
